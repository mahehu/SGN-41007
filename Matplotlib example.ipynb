# Matplotlib example

In this example, we will draw a simple matplotlib chart using matplotlib.

First import the plotting library and call it "plt" for short.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Prepare an example plot...

```python
fig,ax = plt.subplots(1,1, figsize = [10,10])
ax.axis([-3,13,-3,13])
c = plt.Circle(xy = (3,7), radius = 4, alpha = 0.4, fc = 'blue')
ax.add_patch(c)
fontsize = 15
ax.text(1.5, 7.5, 'Hacker\nSkills', ha = 'center', va = 'center', fontsize = fontsize);

c = plt.Circle(xy = (7,7), radius = 4, alpha = 0.4, fc = 'red')
ax.add_patch(c)
ax.text(9, 7.5, 'Substance', ha = 'center', va = 'center', fontsize = fontsize)

y = 7 - 2*np.sqrt(3)
c = plt.Circle(xy = (5, y), radius = 4, alpha = 0.4, fc = 'green')
ax.add_patch(c)
ax.text(5, 1.5, 'Math &\n Statistics', ha = 'center', va = 'center', fontsize = fontsize)

ax.annotate("Danger Zone", 
             xy = (5, 9), 
             xytext=(6, 11.5), 
             size=fontsize,
             bbox=dict(boxstyle="round4", fc="w", ec = "g"),
             arrowprops=dict(arrowstyle="simple",
             connectionstyle="arc3,rad=0.2",
                             shrinkA = 0,
                             shrinkB = 8,
                             fc = "g",
                             ec = "g"),
             horizontalalignment='center', 
             verticalalignment='middle')

ax.annotate("Model Based Research\n(Biology, Physics,...)", 
             xy = (8, 4), 
             xytext=(11, 0), 
             size=fontsize,
             bbox=dict(boxstyle="round4", fc="w", ec = "g"),
             arrowprops=dict(arrowstyle="simple",
                             connectionstyle="arc3,rad=0.2",
                             shrinkA = 0,
                             shrinkB = 8,
                             fc = "g",
                             ec = "g"),
             horizontalalignment='center', 
             verticalalignment='middle')

ax.annotate("Machine Learning", 
             xy = (2, 4), 
             xytext=(-1, 0), 
             size=fontsize,
             bbox=dict(boxstyle="round4", fc="w", ec = "g"),
             arrowprops=dict(arrowstyle="simple",
                             connectionstyle="arc3,rad=-0.2",
                             shrinkA = 0,
                             shrinkB = 8,
                             fc = "g",
                             ec = "g"),
             horizontalalignment='center', 
             verticalalignment='middle')
ax.text(5, 6, 'Superman', ha = 'center', va = 'center', fontsize = fontsize)

ax.axis('equal')
ax.axis('off');
```

That's it: we have created a diagram of different components to data science.
